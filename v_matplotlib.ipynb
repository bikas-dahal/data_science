### StatsModel

```python
%pip install statsmodel
```

```python
import statsmodels.formula.api as smf 
```

```python
avg_by_year.head()
```

```python
mpg_model = smf.ols(
    'comb08 ~ year',
    data=avg_by_year
).fit()
```

```python
mpg_model.fittedvalues
```

```python
mpg_model.summary()
```

```python
fig, ax = plt.subplots() 
ax.scatter(
    x = avg_by_year['year'],
    y = avg_by_year['comb08']
)
ax.plot(
    avg_by_year['year'],
    mpg_model.fittedvalues,
)
```

```python
fig, ax = plt.subplots() 
ax.scatter(
    x = avg_by_year['year'],
    y = avg_by_year['comb08']
)
ax.plot(
    avg_by_year['year'],
    mpg_model.fittedvalues,
)

ax.set_xlim(2010, 2020)
ax.set_ylim(20, 30)
```
