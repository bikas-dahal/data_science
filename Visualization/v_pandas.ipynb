```python
import pandas as pd 
from pathlib import Path
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import ticker
```

```python
src_file = Path.cwd() / 'vehicles.csv'
img_dir = Path.cwd() / 'images'
```

```python
df = pd.read_csv(src_file)
df.head()
```

```python
df['comb08'].plot(kind='hist')
```

```python
df['comb08'].plot(kind = 'kde')
```

```python
df['comb08'].plot(kind = 'box')
```

```python
avg_by_year = df.groupby(['year']).agg(
    {'highway08' : 'mean',
    'city08' : 'mean',
    'comb08' : 'mean'}
)

avg_by_year.head()
```

```python
avg_by_year.plot.box(title = 'Gas Milage')
```

```python
avg_by_year.plot(kind='line', title='Gas Mileage')
```

```python
avg_by_year.plot(kind='line', title='Gas Mileage', figsize=(9, 5),
                 xticks=np.arange(1984, 2026, 5), ylim=(15, 32), rot=45)
```

```python
avg_by_year.plot(kind='barh', rot=45, title='Gas Mileage')
```

```python
avg_by_year.plot(kind='area', title='Gas Mileage')
```

```python
fig, ax = plt.subplots(figsize = (9, 4))
avg_by_year.plot(
    kind = 'area',
    ax= ax,
    stacked=False
)
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:0.0f}'))
ax.set(
    ylabel = 'MPG',
    title = 'Gas Mileage'
)
```

```python
df.boxplot(
    column=['highway08', 'city08', 'comb08'],
    by = 'cylinders', figsize= (12, 5), layout=(1, 3),
    showfliers = False
)
```

```python
from pandas.plotting import scatter_matrix, andrews_curves, parallel_coordinates, radviz
```

```python
df.head()
```

```python
scatter_matrix(df[['highway08', 'city08', 'comb08']], alpha=0.2, figsize=(12, 12))
```

```python
df['VClass'].value_counts()
```

```python
car_class = ['Compact Cars', 'Midsize Cars', 'Subcompact Cars', 'Large Cars']
car_class_df = df[['cylinders', 'fuelCost08', 'co2', 'VClass']].query(
    'VClass == @car_class'
).copy()
```

```python
car_class_df.head()
```

```python
scatter_matrix(car_class_df, alpha=0.2, figsize=(12, 12))
```

```python
fig, ax1 = plt.subplots(figsize = (10, 4))
andrews_curves(car_class_df, 'VClass', ax=ax1)
ax1.legend(loc = 'upper left')
```

```python
fig, ax1 = plt.subplots(figsize = (10, 4))
parallel_coordinates(car_class_df, 'VClass', ax=ax1)
```

```python
fig, ax1 = plt.subplots(figsize = (10, 4))
radviz(car_class_df, 'VClass', ax=ax1)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize = (10, 10), nrows=3, ncols=1)
andrews_curves(car_class_df, 'VClass', ax=ax1)
ax1.legend(loc = 'upper left')
parallel_coordinates(car_class_df, 'VClass', ax=ax2)
radviz(car_class_df, 'VClass', ax=ax3)
```
