```python
import pandas as pd 
from pathlib import Path
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import ticker
import seaborn as sns 
import altair as alt
```

```python
src_file = Path.cwd() / 'vehicles.csv'
img_dir = Path.cwd() / 'images'
```

```python
df = pd.read_csv(src_file)
df.tail()
```

```python
alt.data_transformers.names()
```

```python
alt.data_transformers.enable('json')
```

```python
alt.Chart(df).mark_circle().encode(
    x = 'displ',
    y = 'comb08'
)
```

```python
alt.Chart(df).mark_point().encode(
    x = 'displ',
    y = 'comb08'
)
```

```python
df_edit = df.head(1000)
```

```python
alt.Chart(df_edit).mark_point().encode(
    x = 'displ',
    y = 'comb08',
    color='drive',
    shape='drive'
)
```

```python
alt.Chart(df_edit).mark_bar().encode(
    y = 'count()',
    x = 'comb08',
)
```

```python
df_edit.columns
```

```python
alt.Chart(df_edit).mark_tick().encode(
    y = 'fuelType',
    x = 'barrels08',
)
```

```python
alt.Chart(df_edit).mark_boxplot().encode(
    y = 'fuelCost08',
    x = 'year',
)
```

```python
alt.Chart(df_edit).mark_boxplot().encode(
    y = 'fuelCost08:Q',
    x = 'year:O',
)
```

```python
alt.Chart(df_edit).mark_bar().encode(
    x = 'mean(fuelCost08)',
    y = 'year:O'
)
```

```python
alt.Chart(df_edit).mark_bar().encode(
    alt.X('comb08', type='quantitative', bin=True),
    alt.Y(aggregate='count', type='quantitative'),
)
```

```python
alt.Chart(df_edit).mark_bar().encode(
    alt.X('comb08', bin=alt.Bin(extent = [0, 5000], step = 250)),
    alt.Y(aggregate='count', type='quantitative'),
)
```

```python
alt.Chart(df_edit).mark_point().encode(
    alt.X('displ', type='quantitative'),
    alt.Y('comb08'),
    alt.Color('cylinders', type='ordinal')
)
```

```python
alt.Chart(df_edit).mark_point().encode(
    alt.X('displ', type='quantitative'),
    alt.Y('comb08'),
    alt.Color('cylinders', type='quantitative')
)
```

```python
alt.Chart(df_edit).mark_point().encode(
    alt.X('displ', type='quantitative'),
    alt.Y('comb08'),
    alt.Color('cylinders', type='nominal')
)
```

```python
alt.Chart(df_edit).mark_circle(size=50).encode(
    x = 'displ',
    y = 'comb08',
    tooltip= ['make', 'model', 'year']
).interactive()
```

```python
chart1 = alt.Chart(df_edit).mark_tick().encode(
    y = 'fuelType',
    x = 'barrelsA08'
)

chart2 = alt.Chart(df_edit).mark_bar().encode(
    alt.X('barrels08:Q', bin = True),
    alt.Y('count()')
)

chart1 | chart2
```

```python
chart1 & chart2
```

```python
alt.vconcat(chart1, chart2)
```

```python
alt.Chart(df_edit).mark_circle(size=50).encode(
    x = 'displ',
    y = 'comb08',
    tooltip = ['make', 'model', 'year']
).facet('year:N', columns=2)
```

```python
bars = alt.Chart(df_edit).mark_bar().encode(
    x = 'mean(comb08):Q',
    y = 'year:O'
)

bars
```

```python
x = alt.Chart(df_edit).mark_rule(color='red').encode(
    x = 'mean(comb08):Q'
)

bars + x
```

```python
text = bars.mark_text(
    align='left',
    dx = 3
).encode(text = alt.Text('mean(comb08):Q', format=',.of'))

(bars + x + text)
```
